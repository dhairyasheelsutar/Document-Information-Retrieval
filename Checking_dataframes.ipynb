```python
import joblib
```

```python
bert_embeddings = joblib.load("./models/bert_embeddings.pkl")
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./datasets/keywords.csv")
```

```python
df.head()
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df.head()
```

```python
bert_embeddings.head()
```

```python
bert_embeddings["project"] = df["project"]
bert_embeddings["keyword"] = df["keyword"]
bert_embeddings["file"] = df["file"]
```

```python
bert_embeddings.head()
```

```python
document_topic_feature_matrix = joblib.load("./models/document_topic_feature_matrix.pkl")
```

```python
document_topic_feature_matrix.head()
```

```python
document_topic_feature_matrix.columns
```

```python
columns_order = ["project", "file", "Dominant_Topic", "Topic_Perc_Contrib"]
```

```python
document_topic_feature_matrix = document_topic_feature_matrix[columns_order]
```

```python
document_topic_feature_matrix.head()
```

```python
context_feature_matrix = pd.merge(document_topic_feature_matrix, bert_embeddings, left_on="project", right_on="project")
```

```python
context_feature_matrix.drop("file_x", axis=1, inplace=True)
```

```python
joblib.dump(context_feature_matrix, "./models/context_based_feature_matrix.pkl")
```

```python
joblib.dump(document_topic_feature_matrix, "./models/document_topic_feature_matrix.pkl")
```

```python
joblib.dump(bert_embeddings, "./models/bert_embeddings.pkl")
```
